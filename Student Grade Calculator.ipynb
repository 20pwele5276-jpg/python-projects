Student Grade Calculator

```python
# Student Grade Calculator

# Ask for number of subjects
num_subjects = int(input("Enter number of subjects: "))

total = 0

# Take marks for each subject
for i in range(1, num_subjects + 1):
    marks = float(input(f"Enter marks for subject {i}: "))
    total += marks  # add marks to total

# Calculate average
average = total / num_subjects

# Decide grade using if-elif
if average >= 90:
    grade = 'A'
elif average >= 80:
    grade = 'B'
elif average >= 70:
    grade = 'C'
elif average >= 60:
    grade = 'D'
else:
    grade = 'F'

# Use ternary operator for pass/fail
result = "Pass" if average >= 50 else "Fail"

# Display results
print("\nTotal Marks =", total)
print("Average =", average)
print("Grade =", grade)
print("Result =", result)
```
